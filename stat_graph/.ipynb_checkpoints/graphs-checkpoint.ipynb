```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import matplotlib.dates as mdates
```

```python
custom_datetime = lambda x: datetime.strptime(x, '%d/%b/%Y:%H:%M:%S %z')
rawIntensity = pd.read_csv(
    'u05smevosb02_srv01_access.log.gz_csv',
    delimiter=';',
    parse_dates=[1, ],
    date_parser=custom_datetime,
    names=['IP', 'Date', 'Method', 'Endpoint', 'Protocol', 'response_code', 'Count'])

rawIntensity['Date'] = pd.to_datetime(rawIntensity['Date']).apply(lambda t: t.strftime('%Y-%m-%d %H:%M'))
rawIntensity.head()
```

```python
df = rawIntensity[(rawIntensity.Date > '2015-02-06 00:00') & (rawIntensity.Date < '2015-02-09 00:00')]
df1 = df.groupby(['Date']).count()[['Count']].reset_index()
# print(df1)
dfAllRequests = pd.DataFrame({'Time': df1.Date, 'Count': df1.Count})
dfAllRequests.reset_index(inplace=True, drop=True)
dfAllRequests['Time'] = pd.to_datetime(dfAllRequests['Time']).apply(lambda t: t.strftime('%H:%M'))
```

```python
#   
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(dfAllRequests.Time, dfAllRequests.Count, 'deepskyblue')

xAxis = np.arange(0, dfAllRequests.index[-1],5)
plt.xlim(dfAllRequests['Time'][0], dfAllRequests['Time'][dfAllRequests.index[-1]])
plt.xticks(xAxis, dfAllRequests['Time'][xAxis], rotation='vertical')

ax.set_xlabel(' , :', x=0.85, fontsize=14, fontweight='bold')
ax.set_ylabel('  ', rotation=0, y=1.05, fontsize=14, fontweight='bold')
ax.set_title('', fontsize=18, x=0.9)
ax.grid(axis='y')

plt.show()
plt.close()
```

```python
endpoints = set(df.Endpoint)
print(endpoints)
```

```python
for x in endpoints:
    print(x)
    df3 = df[df.Endpoint == x].groupby(['Date']).count()[['Count']].reset_index()
#     print(df3)
    dfRequest = pd.DataFrame({'Time': df1.Date, 'Count': df3.Count}).fillna(0)
    dfRequest.reset_index(inplace=True, drop=True)
#     print(dfRequest)
    dfRequest['Time'] = pd.to_datetime(dfRequest['Time']).apply(lambda t: t.strftime('%H:%M'))

    #    
    fig, ax = plt.subplots(figsize=(14, 7))
    ax.plot(dfRequest.Time, dfRequest.Count, 'darkviolet')


    xAxis = np.arange(0, dfRequest.index[-1],5)
    plt.xlim(dfRequest['Time'][0], dfRequest['Time'][dfRequest.index[-1]])
    plt.xticks(xAxis, dfRequest['Time'][xAxis], rotation='vertical')

    ax.set_xlabel(' , :', x=0.85, fontsize=14, fontweight='bold')
    ax.set_ylabel('  ', rotation=0, y=1.05, fontsize=14, fontweight='bold')
    ax.set_title('', fontsize=18, x=0.9)
    ax.grid(axis='y')

    plt.show()
    plt.close()
```
